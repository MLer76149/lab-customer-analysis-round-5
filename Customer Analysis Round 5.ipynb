X-y split.

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_numerical.drop(columns=["Total Claim Amount"])
y = data_numerical[["Total Claim Amount"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

Normalize (numerical).

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
MinMaxtransformer = MinMaxScaler()
MinMaxtransformer.fit(X)
x_normalized_min_max = MinMaxtransformer.transform(X)
x_normalized_min_max = pd.DataFrame(x_normalized_min_max,columns=X.columns)
x_normalized_min_max.head()
```

```python
display(x_normalized_min_max.describe().T)
```
